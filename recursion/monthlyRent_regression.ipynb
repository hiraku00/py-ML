実際に使用するのは、都道府県を除いた7つ(都道府県はラベル)<br>
今回家賃の予測を行うが、こういった予測対象の変数を「目的変数」と呼び、<br>
それ以外の6つの変数のことを「説明変数」と呼ぶ<br>
</div>

## データについて
<div style="font-size:14px; color:white; margin:20px">
家賃 : 厳密には民営賃貸住宅の家賃で1カ月3.3$m^2$当たりの価格<br>
人口密度 : 可住地面積1$km2$当たりの人口密度<br>
県内総生産額、事業所数 : 仕事に関する指標として抽出、県内総生産額は単位が億円<br>
コンビニ数、飲食店数、病院数 : 生活するうえでの利便性に関する指標として抽出<br>
</div>

```python
import pandas as pd
allData = pd.read_csv('monthlyRent.csv')
allData
```

<div style="font-size:14px; color:green; margin:20px">
まずは、いくつかある説明変数の内、「人口密度」のみ用いて、家賃を予測
</div>

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(allData['人口密度'], allData['家賃'])
plt.xlabel('populationDensity')
plt.ylabel('MonthlyRent')
```

<div style="font-size:14px; color:green; margin:20px">
このグラフを見ると、人口密度が高くなると家賃も上がる(右肩上がり)傾向にあることがわかる<br>
回帰とは、この右肩上がりの傾向を最もよく表現する線を引くこと(=学習)で、特に直線の場合は線形回帰と言う
</div>

```python
# 機械学習ライブラリscikit-learn(sklearn)の中から、LinearRegression(線形回帰)を呼び出し
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
# 説明変数として人口密度、目的変数として家賃を定義
X1 = allData[['人口密度']]
Y1 = allData['家賃']
# 線形回帰の実行
model1.fit(X1, Y1)
```

<div style="font-size:14px; color:green; margin:20px">
先ほどの散布図に、今回予測した線を追加して可視化
</div>

```python
plt.scatter(allData['人口密度'], allData['家賃'])
plt.xlabel('populationDensity')
plt.ylabel('MonthlyRent')
# Add
plt.plot(X1, model1.predict(X1))
```

<div style="font-size:14px; color:green; margin:20px">
線形回帰における学習とは、既知の家賃、人口密度データを教師データとして用いて、最も誤差の少ない直線を導き出すこと<br>
この直線によって、人口密度から家賃を予測することが可能となった。
</div>

<div style="font-size:14px; color:green; margin:20px">
人口密度が4000の場合の家賃を予測
</div>

```python
model1.predict(4000)
```

<div style="font-size:14px; color:green; margin:20px">
この直線の数式での表現と、モデルの評価
</div>

```python
print(model1.coef_)
print(model1.intercept_)
print(model1.score(X1, Y1))
```

<div style="font-size:14px; color:green; margin:20px">
1行目で傾き、2行目で切片を出力しています。<br>
つまり、家賃は「人口密度 × 0.5 + 3681」で表現できるということを表しており、<br>
この値を教師データから機械に学習
</div>

<div style="font-size:14px; color:green; margin:20px">
あらかじめ家賃がわかっている教師データから予測<br>
そのため、今回学習によって作られた直線が正しいかどうかを評価することが可能<br>
最も単純に見る方法として、決定係数を見る方法がある<br>
決定係数は1に近づくほど、精度が高いことを表しており、<br>
今回のケースの0.769は比較的高い値で、人口密度を用いて家賃を上手く表現できていることを意味してる<br>
ただし、実際には決定係数だけでなく、残差プロット、平均2乗誤差、過学習に陥っていないか等も見る必要がある
</div>

### 変数を拡張して線形回帰
<div style="font-size:14px; color:green; margin:20px">
これまでは、説明変数として人口密度のみを用いて線形回帰を実施<br>
しかし、これ以外にも変数を拡張することで、さらに上手に家賃を表現できる可能性がある<br>
そこで、家賃を目的変数に、それ以外の6項目を説明変数にして回帰を実行<br>
先ほどのように、説明変数が1つの場合は単回帰、複数ある場合は重回帰と呼ぶ
</div>

```python
# 目的変数として家賃を作成
Y2 = allData['家賃']
# 都道府県、家賃の列を削除した説明変数X2を作成
X2 = allData.drop(['都道府県', '家賃'], axis=1)
# データ確認
X2.head()
```

```python
model2 = LinearRegression()
# fitによって、6つの変数と家賃を用いて線形回帰を実行
# 今回は、説明変数が6つあるため、係数も6つ存在
# Y=a1×[人口密度] + a2×[県内総生産額] + a3×[事業所数] + 
# a4×[コンビニ数] + a5×[飲食店数] + a6×病院数 + bとなる
model2.fit(X2, Y2)
# 6つの係数と係数名を一覧で表示
print(pd.DataFrame({'係数名':X2.columns, '係数値':model2.coef_}))
# 切片
print(model2.intercept_)
# 決定係数
print(model2.score(X2, Y2))
```

<div style="font-size:14px; color:green; margin:20px">
決定係数もわずかに増加(0.76 -> 0.79)しており、若干モデルの精度が向上
</div>

<div style="font-size:14px; color:green; margin:20px">
未知のデータから家賃の予測<br>
たとえば、
人口密度4000、国内総生産額50000、事業所数100000、コンビニ数1000、飲食店数14000、病院数140<br>
の場合に家賃がいくらになるか？
</div>

```python
X_pred = pd.DataFrame([[4000, 50000, 100000, 1000, 14000, 140]])
model2.predict(X_pred)
```
