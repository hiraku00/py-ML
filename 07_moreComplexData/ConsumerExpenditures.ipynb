<div style="font-size:20px; color:pink; margin:20px">
データの準備
</div>
<div style="font-size:14px; color:white; margin:20px">
それでは、機械学習のために、データ準備をしていきましょう。
先ほど調べてわかったかと思いますが、
目的変数は消費支出増加フラグ、説明変数は食料から諸雑費、1カ月前から6カ月前までのデータとなります
</div>

```python
# 読み込んだデータを残すために、data_tmpを作成
data_tmp = data.copy()
# 不要なデータを削除
del data_tmp['Date']
del data_tmp['消費支出']
# 説明変数data_tmp_X、目的変数data_tmp_Yを作成
data_tmp_X = data_tmp.copy()
del data_tmp_X['消費支出増加フラグ']
data_tmp_Y = data_tmp['消費支出増加フラグ']
# datatmp_X.columnsで、列がしっかり用意できているか確認
data_tmp_X.columns
```

```python
# データを訓練データと、テストデータに分割
# 分割には、traintestsplitを使用
from sklearn.model_selection import train_test_split
# サイズは、訓練70%、テスト30%に分割
X_train, X_test, Y_train, Y_test = train_test_split(data_tmp_X, data_tmp_Y, random_state=0, test_size=0.3)
# 分割できているか確かめるために、lenを用いてデータ数を単純に合算し、printで出力
print(len(X_train))
print(len(Y_train))
print(len(X_test))
print(len(Y_test))
```

<div style="font-size:20px; color:pink; margin:20px">
決定木
</div>
<div style="font-size:14px; color:white; margin:20px">
まずは、決定木による予測モデル作成をやっていきましょう。
</div>

```python
# 決定木のインポート
from sklearn.tree import DecisionTreeClassifier
# 予測モデルの定義(今回の木の深さは、とりあえず3)
treeModel = DecisionTreeClassifier(max_depth=3, random_state=0)
# 予測モデルを作成
treeModel.fit(X_train, Y_train)
# 精度を把握するために、accuracy_scoreで最も簡易的なモデル評価(精度)を行った
from sklearn import metrics
print(metrics.accuracy_score(Y_train, treeModel.predict(X_train)))
print(metrics.accuracy_score(Y_test, treeModel.predict(X_test)))
```

<div style="font-size:14px; color:white; margin:20px">
訓練データで78%、テストデータで60%となりました。<br>
訓練データでの精度が高いため、訓練データに過剰適合(過学習)傾向にありますが、<br>
訓練、テストデータどちらを用いても60%以上の精度は出ています。<br>
<br>
過学習傾向なため、モデルの複雑さを減らしてみましょう。<br>
モデルをシンプルにする方法として、max_depthを小さくする方法があります
</div>

```python
treeModel = DecisionTreeClassifier(max_depth=2, random_state=0)
treeModel.fit(X_train, Y_train)
print(metrics.accuracy_score(Y_train, treeModel.predict(X_train)))
print(metrics.accuracy_score(Y_test, treeModel.predict(X_test)))
```

<div style="font-size:14px; color:white; margin:20px">
max_depthを2にすると、訓練データの精度は減少し、テストデータの精度は向上しました。<br>
これによって、精度は下がりましたが、モデルは汎化され、より良いモデルになったと考えられます。<br>
最後に、モデルに寄与している変数を抽出してみましょう。<br>
</div>

```python
Importance = pd.DataFrame({'変数名':data_tmp_X.columns, '重要度':treeModel.feature_importances_})
Importance[Importance['重要度']!=0]
```

<div style="font-size:14px; color:white; margin:20px">
これだけ見ると、すべて衣類関係の物価が寄与しているという結果になりました。<br>
解釈は難しいところですが、衣類関連だけですべての消費支出を決めているとは考えにくいです。<br>
まだモデル精度が低いことも考慮する必要があると思います。<br>
しかし、このデータセットを使用する限り、これ以上決定木のモデル精度が向上させるのは困難なため、<br>
ロジスティック回帰に挑戦し、モデル精度を比較してみましょう。<br>
</div>

<div style="font-size:20px; color:pink; margin:20px">
ロジスティック回帰
</div>
<div style="font-size:14px; color:white; margin:20px">
使用するデータは、決定木と全く同じで良いので、訓練データ、テストデータに分割されているものを使用
</div>

```python
# ロジスティック回帰のインポート
from sklearn.linear_model import LogisticRegression
# 予測モデルの定義(決定木と異なり、特にパラメータ指定はせず、デフォルトのものを使用)
logModel = LogisticRegression()
# Fitで予測モデルを作成
logModel.fit(X_train, Y_train)
# 精度評価(簡易的)
print(metrics.accuracy_score(Y_train, logModel.predict(X_train)))
print(metrics.accuracy_score(Y_test, logModel.predict(X_test)))
```

<div style="font-size:14px; color:white; margin:20px">
訓練データで83%、テストデータでも79%を示しました。<br>
決定木よりも全体的な精度が高いだけでなく、訓練データとテストデータの精度に開きがないため、<br>
汎化されている良いモデルとなっています<br>
<br>
前回述べたように、手法によって分類の方法が異なるため、今回のケースのように大きくモデル精度が異なることがあります。<br>
いくつかの手法を用いて適切に評価を行い、最適なモデルを作成していくことの重要性を理解できたのではないでしょうか。<br>
<br>
せっかくなので、ロジスティック回帰のモデルについて、もう少し触れていきましょう。<br>
前回、ロジスティック回帰は、決定木のように条件分岐によって線を引いていくのではなく、<br>
あくまでも多変数の式を元に分割線を引くものであるとお伝えしました。<br>
ロジスティック回帰は、線形回帰の関数Y=aX1+bX2+zを0から1の範囲に押し込めてしまう関数を使用しています。<br>
そのため、線形回帰と同様にモデルを作成することは線を引くことであり、線を引くことは係数(a、b、z)を決めることと同義です。<br>
<br>
ロジスティック回帰には、決定木のようにfeature_importancesが存在しませんが、<br>
係数によってある程度変数の寄与がわかるようになっています。早速、出力してみましょう。<br>
</div>

```python
Coef = pd.DataFrame({'変数名':data_tmp_X.columns, '係数':logModel.coef_[0]})
Coef.sort_values('係数')
```

<div style="font-size:14px; color:white; margin:20px">
1行目で、今回作成したモデルから、係数を抽出しています。<br>
これらの係数は、Y=aX1+bX2+zのaやbに相当します。<br>
つまり、それぞれの変数に掛かっており、絶対値が大きければ大きいほど寄与は大きく、<br>
正であれば支出増加に、負であれば支出減少に寄与していることになります。<br>
絶対値が大きい部分には、1カ月前の物価指数は存在しておらず、<br>
1カ月前の物価指数の寄与は小さいことが考えられます。<br>
つまり、支出増減には少なくとも2カ月以上前の物価が大きく影響していると予想できます。<br>
ただし、住宅の物価指数が、正にも負にも入っており、景気の波等が関係している可能性があります。<br>
さらにより良いモデルにしていく場合、データの粒度や説明変数の見直しも必要になると思います。<br>
</div>
